````python
def gcd(a,b):
    if a == 0:
        return b
    elif b == 0:
        return a
    else:
        return gcd(b,a%b)
````

# filter

````python
nums = [10,24,56,31,25,9,78]
even = list(filter(lambda a:a%2==0,nums))
even
````

````python
def even_fun(n):
    if n%2==0:
        return True
    else:
        False
evn = list(filter(even_fun,nums))
evn
````
